# roboschool

```python
roboschool_env_names = ['RoboschoolInvertedPendulum-v1', 'RoboschoolInvertedPendulumSwingup-v1',
                        'RoboschoolInvertedDoublePendulum-v1','RoboschoolReacher-v1',
                        'RoboschoolHopper-v1','RoboschoolWalker2d-v1',
                        'RoboschoolHalfCheetah-v1','RoboschoolAnt-v1', 
                        'RoboschoolHumanoid-v1', 'RoboschoolHumanoidFlagrun-v1', 
                        'RoboschoolHumanoidFlagrunHarder-v1', 'RoboschoolAtlasForwardWalk-v1',
                        'RoboschoolPong-v1']
action_dims = []
for env_name in roboschool_env_names:
    env = gym.make(env_name)
    print('Environment: {}'.format(env_name))
    print('observation_space: {}'.format(env.observation_space))
    print('action_space: {}'.format(env.action_space))
    print('env._max_episode_steps={}'.format(env._max_episode_steps))
    action_dims.append(env.action_space.shape[0])

import numpy as np
action_dims = np.array(action_dims)
np.unique(action_dims)
```

# MuJoCo
Continuous control tasks, running in a fast physics simulator.

* **Ant-v2**
  * observation_space: **Box(111,)**
  * action_space: **Box(8,)**
* **HalfCheetah-v2**
  * observation_space: **Box(17,)**
  * action_space: **Box(6,)**
* **Hopper-v2**
  * observation_space:**Box(11,)**
  * action_space: **Box(3,)**
* **Humanoid-v2**
  * observation_space: **Box(376,)**
  * action_space: **Box(17,)**
* **HumanoidStandup-v2**
  * observation_space: **Box(376,)**
  * action_space: **Box(17,)**
* **InvertedDoublePendulum-v2**
  * observation_space: **Box(11,)**
  * action_space: **Box(1,)**
* **InvertedPendulum-v2**
  * observation_space: **Box(4,)**
  * action_space: **Box(1,)**
* **Reacher-v2**
  * observation_space: **Box(11,)**
  * action_space: **Box(2,)**
* **Swimmer-v2**
  * observation_space: **Box(8,)**
  * action_space: **Box(2,)**
* **Walker2d-v2**
  * observation_space: **Box(17,)**
  * action_space: **Box(6,)**

```python
MuJoCo_env_names = ['Ant-v2', 'HalfCheetah-v2', 'Hopper-v2', 'Humanoid-v2', 
                    'HumanoidStandup-v2', 'InvertedDoublePendulum-v2', 'InvertedPendulum-v2', 'Reacher-v2',
                    'Swimmer-v2', 'Walker2d-v2']

def print_env_info(env_names):
    action_dims = []
    for env_name in env_names:
        env = gym.make(env_name)
        print('Environment: {}'.format(env_name))
        print('  observation_space: {}'.format(env.observation_space))
        print('  action_space: {}'.format(env.action_space))
        if type(env.action_space) ==gym.spaces.box.Box: #gym.spaces.discrete.Discrete:
            print('  action_space.shape = {}'.format(env.action_space.shape))
            print('  action_space.high = {}'.format(env.action_space.high))
            action_dims.append(env.action_space.shape[0])
        print('  env._max_episode_steps={}'.format(env._max_episode_steps))
        print('\n')
        
    return action_dims

action_dims = print_env_info(MuJoCo_env_names)

import numpy as np
action_dims = np.array(action_dims)
np.unique(action_dims)
```
